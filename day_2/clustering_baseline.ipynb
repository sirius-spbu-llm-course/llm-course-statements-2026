```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

# Загружаем данные
# В этом файле нет колонки 'species' - мы должны её угадать!
df = pd.read_csv('datasets/penguins_X.csv')

print(f"{df.shape[0]} rows")
df.head()
```

```python
# Выбор признаков

features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
X = df[features]

X = X.fillna(0) # Заменили все пропуски на 0. Норм идея? Или пингвинов весом 0 кг не бывает

# Что-то еще сюда нужно
```

```python
model = KMeans(n_clusters=3, random_state=42)

labels = model.fit_predict(X)

# Можно посмотреть такую штуку
print("Сколько пингвинов в каждом кластере:", np.bincount(labels))
```

```python
# Генерация решения

submission = pd.DataFrame({
    'id': range(len(labels)),
    'species': labels
})

submission.to_csv('submission.csv', index=False)
```
